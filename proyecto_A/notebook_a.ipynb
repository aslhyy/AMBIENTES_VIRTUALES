# Promedio de Notas con Visualización

En este notebook calculamos el **promedio de notas de varios estudiantes**
y mostramos un gráfico de barras con los resultados.


```python
# Diccionario de estudiantes con sus notas
notas = {
    "Ana": [4.2, 3.8, 4.5],
    "Luis": [3.0, 3.5, 3.7],
    "Sofía": [4.8, 4.6, 5.0],
    "Carlos": [2.8, 3.1, 3.0]
}

# Calcular promedio de cada estudiante
promedios = {nombre: sum(valores) / len(valores) for nombre, valores in notas.items()}

print("Promedio de notas por estudiante:")
for estudiante, promedio in promedios.items():
    print(f"{estudiante}: {promedio:.2f}")
```

```python
import matplotlib.pyplot as plt

nombres = list(promedios.keys())
valores = list(promedios.values())

plt.bar(nombres, valores, color=["#6a5acd", "#6495ed", "#00ced1", "#ffa07a"])
plt.title("Promedio de Notas por Estudiante")
plt.xlabel("Estudiantes")
plt.ylabel("Promedio")
plt.ylim(0, 5)
plt.show()
```
